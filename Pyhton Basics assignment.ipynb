```python
# Q3. Compare and contrast mutable and immutable objects in Python with examples.

#ANS: Objects are classified as either mutable or immutable based on whether their state (i.e., their data) 
# can be changed after they are created.

# Mutable Objects: Mutable objects can be changed after they are created.
#    This means you can modify their contents or state without creating a new object.
# example: Lists: Lists in Python are mutable.You can change, add, or remove elements after the list is created.

# Creating a list
my_list = [1, 2, 3]
print(my_list) 

#modifying the list 
my_list.append(4)
print(my_list)

my_list[1] = 99
print(my_list)

 #example 2 : Dictionaries: Dictionaries are mutable. You can change, add, or remove key-value pairs
    
    # Creating a dictionary
my_dict = {'a': 1, 'b': 2}
print(my_dict)

# Modifying the dictionary
my_dict['c'] = 3
print(my_dict) 

del my_dict['b']
print(my_dict)

#  Characteristics:

#  a. Changes to mutable objects affect the original object.
#  b. Mutable objects are generally more flexible because they allow modification of 
#           their content without the need to create a new object.

# IMMUTABLE OBJECTS:  Immutable objects cannot be changed after they are created.
#                     Any modification results in the creation of a new object 

# Examples:

#  Tuples: Tuples are immutable. Once a tuple is created, its elements cannot be altered.
  # Creating a tuple
my_tuple = (1, 2, 3)
print(my_tuple)

# Attempting to modify the tuple (will raise an error)
# my_tuple[1] = 99  # TypeError: 'tuple' object does not support item assignment.

# Strings: Strings are immutable. Any operation that seems to modify a string actually creates a new string.

# Creating a string
my_string = "hello"
print(my_string)  # Output: hello

# Modifying the string (actually creates a new string)
new_string = my_string.upper()
print(new_string)

# Characteristics:

# Changes to immutable objects result in the creation of a new object with the new value, leaving the original object unchanged.
#  Immutable objects are generally safer in concurrent or multi-threaded environments since their state cannot be altered
```

```python
#  Q 6. How do conditional statements work in Python ? Illustrate with examples.

#  ANS:  Conditional statements in Python allow you to execute certain blocks of code based on whether a condition is true or false.
#       They are essential for making decisions and controlling the flow of a program.

# Python provides several types of conditional statements:
#  Basic Syntax of Conditional Statements
#   The basic syntax of an if statement is:

# a. if condition:
 

#  if condition:
    # Code to execute if the condition is true
#   We can extend this with elif (short for "else if") and else to handle multiple conditions:


#   if condition1:
        # Code to execute if condition1 is true
#   elif condition2:
       # Code to execute if condition1 is false and condition2 is true
#   else:
       # Code to execute if none of the above conditions are true
#   Examples
#    1. Simple if Statement
#     The simplest form of a conditional statement is the if statement:

temperature = 30

if temperature > 25:
    print("It's a hot day!")

    
    
#  2. if and else
#     The else statement provides an alternative block of code that executes if the if condition is false:

temperature = 20

if temperature > 25:
    print("It's a hot day!")
else:
    print("It's a cool day!")


#  3. if, elif, and else
#       When we need to check multiple conditions, you use elif:


temperature = 16

if temperature > 25:
    print("It's a hot day!")
elif temperature > 15:
    print("It's a warm day!")
else:
    print("It's a cool day!")


#   4. Nested if Statements
#      We can nest if statements inside other if statements:

temperature = 30
humidity = 60

if temperature > 25:
    if humidity > 50:
        print("It's a hot and humid day!")
    else:
        print("It's a hot but dry day!")
else:
    print("It's a cool day!")

    
    
#  Combining Conditions
#   We can combine multiple conditions using logical operators:

#   and: Both conditions must be true.
#   or: At least one of the conditions must be true.
#   not: Inverts the boolean value.

temperature = 30
humidity = 45

if temperature > 25 and humidity < 50:
    print("It's a hot day with low humidity.")
else:
    print("The weather conditions are different.")
```

```python
# Q 7. Describe the different types of loops in Python and their use cases with examples.

# ANS:  loops are used to execute a block of code repeatedly under certain conditions.
#       Python supports several types of loops: for loops, while loops, and control flow statements that modify loop behavior.
#        Each type of loop has its own use cases and syntax.


#  1. for Loop
#   The for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range) or other iterable objects.
#   It is often used when the number of iterations is known beforehand or when you need to iterate through a collection of items.


# Syntax:
#              for variable in iterable:
    # Code to execute for each item in the iterable.
    
    
#  Examples:
#            Iterating Over a List

fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)

#     Using range() Function:

for i in range(5):  # Iterates from 0 to 4
    print(i)


#  Iterating Over a String:

for char in "hello":
    print(char)


#  2. while Loop
#    The while loop repeatedly executes a block of code as long as a given condition is True.
#     It is used when the number of iterations is not known in advance and depends on a condition being met.
#   Syntax:
#         while condition:
    # Code to execute while the condition is true.
    
#    Examples:
#     Basic while Loop:

count = 0

while count < 5:
    print(count)
    count += 1
    
#  Using while Loop for User Input:

user_input = ""

while user_input != "exit":
    user_input = input("Type 'exit' to quit: ")
    print(f"You typed: {user_input}")
    
    
    
#  Python provides several statements to control the behavior of loops:

#   break: Exits the loop prematurely.
#   continue: Skips the current iteration and continues with the next iteration of the loop.
#   pass: A placeholder that does nothing; used when a statement is syntactically required but you don't want to execute any code.

#   Examples:
#   Using break:

for num in range(10):
    if num == 5:
        break  # Exits the loop when num is 5
    print(num)

#   Using continue:

for num in range(10):
    if num % 2 == 0:
        continue  # Skips the current iteration for even numbers
    print(num)
    
#    Using pass:

for num in range(5):
    if num % 2 == 0:
        pass  # Does nothing for even numbers
    else:
        print(num)
        
        
#   4. Nested Loops
#   Loops can be nested within other loops.
#   This is useful for iterating over multi-dimensional data structures, like matrices.

#    Example:

#   Nested for Loops:

for i in range(3):  # Outer loop
    for j in range(3):  # Inner loop
        print(f"i={i}, j={j}")
        
#    Nested while Loops:

i = 0
while i < 3:  # Outer loop
    j = 0
    while j < 3:  # Inner loop
        print(f"i={i}, j={j}")
        j += 1
    i += 1
```
